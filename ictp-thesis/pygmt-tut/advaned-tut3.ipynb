```python
import pygmt

grid = pygmt.datasets.load_earth_relief(resolution="10m", region=[-108, -103, 35, 40])

fig = pygmt.Figure()
fig.grdview(
    grid=grid,
    perspective=[130, 30],
    frame=["xa", "yaf", "WSnE"],
    projection="M15c",
    zsize="1.5c",
    surftype="s",
    cmap="geo",
    # Set the plane elevation to 1,000 meters and make the fill "gray"
    plane='"1000+ggray"',
)
fig.show()
```
