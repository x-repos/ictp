```python
import pygmt
```

```python
fig = pygmt.Figure()
fig.basemap(region="g", projection="W15c", frame=True)
fig.coast(shorelines=True)
fig.show()
```

```python
fig = pygmt.Figure()
fig.basemap(region='g', projection='W15c', frame=True)
fig.coast(shorelines=['1/0.5p,black', '2/0.5p,lightblue']) # 0.5 point black line
fig.savefig("test.pdf")
fig.show()
```

```python
oahu = [-158.3, -157.6, 21.2, 21.8]
fig = pygmt.Figure()
for res in ["c", "l", "i", "h", "f"]:
    fig.coast(resolution=res, shorelines="1p", region=oahu, projection="M5c")
    fig.shift_origin(xshift="5c")
fig.show()
```

```python
fig = pygmt.Figure()
fig.basemap(region="g", projection="W15c", frame=True)
fig.coast(land="white", water="skyblue", shorelines=["1/0.5p,black", "2/0.5p,darkblue"])
fig.show()
```
