```python
import pygmt
```

```python
x=[1, 6, 9]
y=[5, 7, 4]

fig = pygmt.Figure()
fig.basemap(
    region=[0, 10, 0, 10],
    projection="X15c/10c",
    frame = ['af','+tPlot']
)
fig.plot(
    x = x,
    y = y,
    pen="1p,black",
)
x=[2, 4, 10]
y=[3, 8, 9]
fig.plot(
    x = x,
    y = y,
    pen="1p,red,-...",
)
fig.show()
```
