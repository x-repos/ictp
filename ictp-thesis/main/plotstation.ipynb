```python
import pandas as pd
import pygmt
data = pd.read_table("coordinate.txt", sep=',')

region = [
    data.longitude.min() - 1,
    data.longitude.max() + 1,
    data.latitude.min() - 1,
    data.latitude.max() + 1,
]

fig = pygmt.Figure()
fig.basemap(region=region, projection="M15c", frame=['afg','+tBanda Arc'])
fig.coast(land="black", water="skyblue")
for i in range(len(data)):
    for j in range(i+1, len(data)):
        fig.plot(
            x=[data.longitude[i], data.longitude[j]],
            y=[data.latitude[i], data.latitude[j]],
            pen="0.5p,white"
            )
fig.plot(
    x=data.longitude,
    y=data.latitude,
    style="i0.4c",
    fill="grey",
    pen="1p,red")
bandasea = [126.2743, -5.6618]
temorsea = [127.7332, -11.475]
arafurasea= [136.0922, -8.8693]
fig.plot(
    x=[121.0887, 133.3494, 133.3494, 121.0887, 121.0887],
    y=[-10.142, -10.142, -1.1007, -1.1007, -10.142],
    pen="1.5p,darkgreen"
    )
fig.text(text=["BANDA SEA", "TEMOR SEA", "ARAFURA SEA"],
         x=[126.2743, 127.7332, 136.0922],
         y=[-5.6618, -11.475, -8.8693])
fig.show()
```
