```python
import os
from glob import glob
import noiseben
import numpy as np
import matplotlib.pyplot as plt
from obspy import read, Trace
from parameters import *
import pandas as pd

# change here for plotting
pathin_stack    = './cc-stack/'
pathin_metadata = './metadata/'
pathout_plot    = './cc-stack-plot/'
sac             = '.sac'
noiseben.makefolder(pathout_plot)

ccpair_path = glob(pathin_stack+'*')
ccpair_path.sort()
ccpair_name  = [os.path.basename(x) for x in glob(pathin_stack+'*')]
ccpair_name.sort()
df1 = pd.read_csv(pathin_metadata+'cc-count.csv', index_col='ccname')
df2 = pd.read_csv(pathin_metadata+'cc-metadata.csv', index_col='ccname')

for i in range(len(ccpair_name)):
    ccpair_name[i] = ccpair_name[i][:len(ccpair_name[i])-len(sac)]

for i in range(len(ccpair_path)):
    print('ploting: '+ccpair_name[i])
    tr = read(ccpair_path[i])[0]
    dt = 1/samp_freq
    tvec = np.linspace(-maxlag,maxlag+dt,len(tr.data))
    plt.figure(figsize=(15,3))
    plt.plot(tvec, tr.data,'black',
             label=ccpair_name[i] +'_'
             +str(df1.loc[ccpair_name[i]]['cc-count'])+'days'+'_'
             +str(round(df2.loc[ccpair_name[i]]['dist']/1000))+'km')
    plt.legend(loc=2)
    plt.axvline(x = 0, color = 'tab:blue', linestyle = ':')
    plt.savefig(pathout_plot+ccpair_name[i]+'.png')
    # plt.show()
    plt.close()
```
