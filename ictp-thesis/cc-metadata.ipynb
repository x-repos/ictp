```python
from obspy import read_inventory
import pandas as pd
from glob import glob
import noiseben
import obspy
import os
from glob import  glob


pathin_stations  = './stations/'
pathout_metadata = './metadata'
noiseben.makefolder(pathout_metadata)
xml = '.xml'
stations_link = glob(pathin_stations+'*')
stations_link.sort()
inv = []
nstation    = len(stations_link)
stationname = [os.path.basename(x) for x in stations_link]
for i in range(len(stationname)):
    stationname[i] = stationname[i][:len(stationname[i])-len(xml)]
for i in range(nstation):
    inv_sub = read_inventory(stations_link[i])
    inv.append(inv_sub)

pathin_cc = './cc/'

ccname = []
ssta = []; snet = []; slon = []; slat = []; selv = []; sloc = []
rsta = []; rnet = []; rlon = []; rlat = []; relv = []; rloc = []
dist = []
for i in range(nstation-1):
    for j in range(i+1,nstation):
        ssta_sub,snet_sub,slon_sub,slat_sub,selv_sub,sloc_sub = noiseben.sta_info_from_inv(inv[i])
        rsta_sub,rnet_sub,rlon_sub,rlat_sub,relv_sub,rloc_sub = noiseben.sta_info_from_inv(inv[j])
        ssta.append(ssta_sub); snet.append(snet_sub); slon.append(slon_sub)
        slat.append(slat_sub); selv.append(selv_sub); sloc.append(sloc_sub)
        dist_sub,_,_ = obspy.geodetics.base.gps2dist_azimuth(slat_sub,slon_sub,rlat_sub,rlon_sub)
        rsta.append(rsta_sub); rnet.append(rnet_sub); rlon.append(rlon_sub);
        rlat.append(rlat_sub); relv.append(relv_sub); rloc.append(rloc_sub)
        dist.append(dist_sub)
        ccname.append(ssta_sub+'_'+rsta_sub)
        
dict_meta = {
    'ccname': ccname,
    'ssta': ssta, 'snet': snet, 'slon': slon, 'slat': slat, 'selv': selv, 'sloc': sloc,
    'rsta': rsta, 'rnet': rnet, 'rlon': rlon, 'rlat': rlat, 'relv': relv, 'rloc': rloc,
    'dist': dist
}

df = pd.DataFrame(dict_meta)
df.to_csv(pathout_metadata+'/'+'cc-metadata.csv', index=False)
```
