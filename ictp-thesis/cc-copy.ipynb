```python
import os
from glob import glob
import noiseben
import numpy as np
import matplotlib.pyplot as plt
from obspy import read, Trace
from parameters import *
import shutil

pathin_cc           = './cc-daily/'
pathout_cc_link     = './cc-link/'
sac                 = '.sac'
txt                 = '.txt'
noiseben.makefolder(pathout_cc_link)

stackorder = 0

# add link for faster processing
print('making link')
time_list = [os.path.basename(x) for x in glob(pathin_cc+'*')]
time_list.sort()
for currenttime_str in time_list:
    cc_name_path = pathin_cc+currenttime_str+'/'
    cc_name_full = glob(cc_name_path+'*')
    cc_name_full.sort()
    cc_name = [os.path.basename(x) for x in cc_name_full]
    for i in range(len(cc_name)):
        cc_name[i] = cc_name[i][0:len(cc_name[i])-11]
        file = open(pathout_cc_link+cc_name[i]+'.txt','a')
        file.writelines(cc_name_full[i]+'\n')
        file.close()


# copy
pathin_cc_link      = './cc-link/'
pathout_cc          = './cc/'
noiseben.makefolder(pathout_cc)

cclink = glob(pathin_cc_link+'*')
cclink.sort()
ccname  = [os.path.basename(x) for x in glob(pathin_cc_link+'*')]
ccname.sort()                             # remember to sort after reading from glob
for i in range(len(ccname)):
    ccname[i] = ccname[i][0:len(ccname[i])-len(txt)]
for i in range(len(cclink)):
    file = open(cclink[i],'r')
    lines = file.readlines()
    noiseben.makefolder(pathout_cc+ccname[i])
    print('copying: '+ccname[i])
    for line in lines:
        line = line[:len(line)-1]
        shutil.copy2(line,pathout_cc+ccname[i])

try:
    shutil.rmtree(pathin_cc_link)
except:
    pass
```
